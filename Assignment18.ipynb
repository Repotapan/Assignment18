Q1)What is Flask Framework? What are the advantages of Flask Framework?

Flask is a web application framework written in Python. It is classified as a micro-framework because it does not require particular tools or libraries, but it still provides the necessary components for building web applications. Flask was created by Armin Ronacher in 2010 and is distributed as an open-source software under a BSD-3-Clause license.

Some advantages of using Flask are:

Lightweight: Flask is a lightweight framework that does not require much setup or configuration. This makes it easy to get started with and flexible enough to fit a variety of use cases.

Flexibility: Flask allows you to add or remove features as needed, which gives you the flexibility to build your web application according to your requirements.

Simplicity: Flask is designed to be simple and easy to use, with a straightforward syntax that makes it easy to learn and understand.

Modular: Flask is a modular framework, which means that it is composed of small and independent components that can be used together or separately. This makes it easy to scale and maintain your web application.

Extensible: Flask provides an extension architecture that allows you to add functionality to your web application easily. There are many third-party extensions available that can help you add features like user authentication, database integration, and more.

Large Community: Flask has a large and active community of developers, which means that there is a wealth of resources available online, including documentation, tutorials, and support forums.

Overall, Flask is a great choice for building small to medium-sized web applications. Its simplicity, flexibility, and extensibility make it a popular choice among Python developers.






Q2)Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook?

```python
from flask import Flask
app = Flask(__name__)
@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3)What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of mapping URLs to view functions in a Flask application. This is done using the @app.route() decorator in Flask. When a user makes a request to a specific URL, Flask checks its defined routes to see which function should handle that request.

We use app routes in Flask for several reasons:

To map URLs to view functions: App routing allows us to define which function should handle a specific URL. This is useful when building a website or web application with multiple pages or sections.

To organize our code: By defining app routes, we can organize our code into smaller, modular functions. This makes it easier to maintain and debug our code.

To handle user input: App routes allow us to handle user input from forms or other sources. For example, we might define an app route that handles form submissions and processes the data that was submitted.

Q5) What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function?

The url_for() function is used to build URLs for a specific function. It takes the name of the function as its first argument and any number of keyword arguments, which represent the variable parts of the URL. Here's an example:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

with app.test_request_context():
    # the following line generates the URL for the hello() function
    url = url_for('hello', name='John')
    print(url) # output: '/hello/John'
```
